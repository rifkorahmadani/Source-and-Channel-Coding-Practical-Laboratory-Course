2. Buat program yang sama dengan diatas tapi menggunakan ```while``` loop

````python
string = "Politeknik Negeri Bandung"
i = 0 #agar syarat "i < len(string):"" nya bisa terpenuhi
while i < len(string):
    if string[i] != " ":
        print(string[i], end="")
    i += 1 #kondisi dimana setiap loop dia akan menambahkan 1 ke nilai variabel i
````

3. Buat program untuk mencetak setiap karakter non-numerik dari "P0liteknik Negeri Bandung 2025"

````python
#buat program untuk counting "huruf_dicari" -> misalnya untuk e itu ada 3 kali kemunculan
#buat program untuk mencetak indeks terakhir -> looping string, lalu ketika bertemu e maka countingnya turun, dan jika countingnya menjadi 0 maka print indeksnya  #countingnya turun ke 0 maka indeksnya diceta

string = 'Politeknik Negeri Bandung'
huruf_dicari= 'e'
i = 0
indeks_akhir = -1

while i < len(string):
    if string[i] == huruf_dicari:
        indeks_akhir = i
    i += 1

print('indeks terakhir huruf e =', + indeks_akhir)
````
