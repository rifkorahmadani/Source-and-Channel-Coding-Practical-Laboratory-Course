### Pendahuluan

Pada modul sebelumnya, kita telah mencoba membuat encoder-decoder kompresi data dengan
menggunakan teknik Run Length Encoding. Pada kali ini, kita akan mencoba membuat encoder-decoder kompresi data yang masih serumpun dengan metode RLE, tetapi dengan detail yang
berbeda, yaitu metode Move to Front. Konsepnya adalah mengupayakan agar simbol-simbol yang
sering muncul ditempatkan pada posisi-posisi awal dengan nomor indeks rendah, sehingga bila
di-encode, cukup menggunakan panjang bit yang sedikit.
Metode ini dimulai dengan sederet simbol dengan posisi setiap simbol direkam dalam indeks. Bila
ada data yang ingin dikirim, maka indeks data tersebut dikirim, dan simbol yang sama pada list
dipindahkan ke posisi terdepan. Begitu seterusnya sampai semua data berhasil dikirim. Sebagai
contoh, bila kita memiliki sederet simbol S = < A, B, C, D, E > dengan indeks posisi I = < 0, 1, 2,
3, 4 >, maka untuk mengirim data D = < D, C, D, E, A, A, B, E, B >, proses yang terjadi dapat
dilihat pada tabel 1.

Pada percobaan pendahuluan ini konsep move to front atau memindahkan simbol ke depan dictionary belum akan diimplementasikan. Fokus dari modul ini, mahasiswa mampu mencetak nilai index datastream input terhadap dictionary yang telah ditetapkan. Adapun langkah yang dapat membantu untuk bisa membuat program tersebut adalah sebagai berikut:

1. Program harus bisa melakukan scan satu persatu terhadap tiap huruf pada datastream secara berurutan dari awal (seperti yang sudah pernah dilakukan sebelumnya). Contoh pada datastream = DCDEAABEB;
program harus bisa mulai mengambil data pertama yaitu huruf D.

Contoh saat datastream membaca huruf D tadi, huruf tersebut dibandingkan dengan dictionary= ABCDE; dengan indexnya 0123. Huruf D akan dibandingkan dengan A, kemudian B, kemudian C, dan kemudian D; saat bertemu huruf yang sama output akan mencetak index D pada dictionary, yaitu index 3.

3. Lanjutkan pembacaan datastream ke index berikutnya. Setelah selesai membaca huruf D, lanjutkan pembacaan ke huruf C. Lakukan step yang sama seperti di atas untuk membandingkan dengan dictionary. Sehingga akan mengeluarkan output index 2. Maka setelah itu output akan menjadi output=32..

4. Lakukan proses tersebut berulang hingga akhir huruf/symbol pada datastream.

**Instruksi:**

Sebagai pendahuluan dari program Move to Front buatlah program untuk bisa mendeteksi angka urutan pada data input terhadap library yang kita buat (deratan simbol).

Deretan symbol yang digunakan adalah:
S = < A, B, C, D, E >

Sedangkan data stream / data input yang digunakan adalah:
datastream = < D, C, D, E, A, A, B, E, B >


a. Insertkan flowchartnya pada laman text di bawah.

b. Tuliskan programnya pada laman code di bawah.

Output yang harus dihasilkan:

datastream = < D, C, D, E, A, A, B, E, B >

output = 323400141



```python
#tuliskan program yang telah dibuat
datastream=['D','C','D','E','A','A','B','E','B'] #data yang mau diubah ke angka
simbol=['A','B','C','D','E'] #daftar acuan huruf A samapai E
matching = '' #wadah kosong buat hasil.

for p in range (len(datastream)): #loop untuk membaca setiap simbol dalam datastream
  for q in range (len(simbol)): #loop untuk mengecek simbol acuan satu per satu
    if datastream[p] == simbol[q]: #Membandingkan simbol input dengan simbol di daftar acuan
      matching = matching + str(q) #Jika cocok, tambahkan posisi indeks simbol tersebut ke hasil

print('datastream =', datastream) #Menampilkan isi datastream
print('output =', matching) #Menampilkan hasil akhir
```



> Analisa dari program di atas:

*Program ini bekerja dengan cara membaca huruf dari `datastream` satu per satu, lalu mencari posisi huruf tersebut pada daftar `simbol=['A','B','C','D','E']`. Setiap kali ada kecocokan, program mengambil nomor indeksnya dan menyusunnya menjadi deretan angka. Misalnya huruf D berada di indeks ke-3, C di indeks ke-2, E di indeks ke-4, dan seterusnya. Dari proses ini, seluruh data input berhasil diubah menjadi hasil angka `323400141`.*

## **==============================opsional/bonus==============================**

1. buat dia insensitif terhadap lowercase atau uppercase ('a' itu samadengan 'A')

```python
datastream = ['d','c','D','E','a','A','B','e','b']
simbol     = ['A','B','C','D','E']
matching   = ''

for p in range(len(datastream)):
    for q in range(len(simbol)):
        if datastream[p].upper() == simbol[q]:   # huruf input diubah ke uppercase
            matching = matching + str(q)

print("datastream =", datastream)
print("output    =", matching)
```

2. dictionary dibuat dari sendiri (array untuk simbol di generate sendiri)

```python
simbol = [chr(i) for i in range(65, 70)]   # 65='A', 69='E'
datastream = ['D','C','D','E','A','A','B','E','B']

matching = ''
for p in range(len(datastream)):
    for q in range(len(simbol)):
        if datastream[p].upper() == simbol[q]:
            matching += str(q)

print("datastream =", datastream)
print("output    =", matching)
```

3. buat datastream dapat diinput oleh user secara manual (menggunakan input())

```python
simbol = [chr(c) for c in range(65, 91)]
datastream = input()
matching = ''

for x in range(len(datastream)):
    for y in range(len(simbol)):
        if datastream[x].upper() == simbol[y].upper():
            matching += str(y)

print("datastream =", datastream)
print("output =", matching)
```
