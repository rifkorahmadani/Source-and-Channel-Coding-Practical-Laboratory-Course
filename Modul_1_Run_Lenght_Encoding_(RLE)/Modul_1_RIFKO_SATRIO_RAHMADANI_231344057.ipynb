## Modul 1: Run Length Encoding (RLE)
## Bagian Encoder

### Pendahuluan

Dalam usaha untuk mengirimkan data dari satu tempat ke tempat lain, seringkali terjadi ketidakseimbangan antara jumlah data yang akan dikirim dengan kapasitas jaringan yang tersedia. Akibatnya, kualitas layanan dalam sistem pengiriman data menjadi kurang memadai. Salah satu cara untuk mengatasi situasi ini adalah dengan melakukan kompresi data sebelum pengiriman, sehingga beban pada jaringan dapat berkurang. Teknik kompresi data dapat diterapkan jika terdapat pola berulang dalam data, di mana pola-pola ini dapat diwakili dengan singkat pada awalnya dan kemunculan selanjutnya hanya merujuk pada pola awal tersebut. Terdapat berbagai teknik kompresi data yang sudah dikembangkan, dan pada kesempatan ini akan diterapkan metode kompresi data yang dikenal sebagai Run Length Encoding (RLE). RLE merupakan teknik kompresi data yang mengompresikan pola yang muncul secara berurutan. Contoh penggunaan RLE adalah dengan mengganti setiap rangkaian angka 0 dengan 2 karakter, yang menunjukkan bahwa kompresi sedang dilakukan, diikuti dengan jumlah angka 0 dalam rangkaian tersebut. Semakin banyak angka 0 yang berurutan, maka ukuran file hasil kompresi akan semakin efisien.

Buktikan program saudara dengan data berikut ini:

**Datastream** = 4831979530008959793400000443500004839859719

**Encoded Data (RLE)**= 4831979530389597934054435044839859719

**Clue:**
Dari program sebelumnya kita sudah bisa menghitung jumlah angka 0 yang berulang. Akan tetapi, hasil dari perhitungan tersebut masih dicetak pada akhir program.

Ubah bagian cetak perhitungan tersebut, agar bisa langsung dicetak setelah angka 0.

Di awal ketika sistem belum bertemu angka 0, maka sistem akan mencetak ulang data yang sedang dibaca.

Namun, saat bertemu angka 0 program akan menghitung selama angka di depannya masih 0.

Di saat angka di depannya bukan angka 0, maka program harus berheti menghitung dan kembali hanya mencetak ulang datastream (angka selain 0).

````python
#Buat Program disini
# input -> Datastream = 383231900387230000398123894189315561000001
# outputnya -> 3832319023872304398123894189315561051

datastream='4831979530008959793400000443500004839859719'
output=''
count=0

for x in range(len(datastream)):
    if datastream[x] == '0':
        count += 1         # hitung deret 0 nya

        if x == len(datastream)-1 or datastream[x+1] != '0':
            output += '0' + str(count)    # flush deret 0
            count = 0                       # me-reset counter
    else:
        output += datastream[x]

# jaga-jaga, kalau berakhir dengan 0
if count > 0:
    output += '0' + str(count)

print(output == '4831979530389597934054435044839859719')
print(output)
````

Dari program tersebut, intinya bagiamana mengkompres deret angka yang berangka 0, conothnya jika ada 0 berderetan sepanjang x, maka saya tulis jadi "0"+k. selain 0 maka ditulis seperti asalnya saja.

Cara kerja programnya,

Output menampung hasil, count menghitung panjang deret 0.

Ke loop baca tiap karakter:
- Kalau ketemu "0" maka count += 1.

- Kalau deret 0 berakhir (karakter berikutnya bukan '0' atau sudah di ujung)
maka tulis ke `output: output += '0' + str(count)` lalu reset `count = 0`

- Kalau bukan "0" maka langsung tambahkan saja ke output.

kemudian untuk program ini ,
```

    output += '0' + str(count)
```
itu sebenarnya opsional setelah loop, kalau masih ada sisa 0 nya
